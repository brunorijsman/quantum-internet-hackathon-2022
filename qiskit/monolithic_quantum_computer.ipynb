```python
from quantum_computer import MonolithicQuantumComputer
from examples import (
    EntanglementExampleCluster,
    TeleportExampleCluster,
    LocalControlledPhaseExampleCluster,
)
from utils import density_matrix_pretty_print
from numpy import pi
```

Create a monolithic quantum computer with four qubits:

```python
total_nr_qubits = 4
computer = MonolithicQuantumComputer(total_nr_qubits)
```

Display the circuit for the quantum computer. Since we have not yet added any gates to the circuit, the only thing we see is the qubits. In the case of a monolithic quantum computers, all qubits are available for user computation (so-called main qubits). In the case of clustered quantum computers, we would also have ancillary qubits, but we don't have any in the case of a monolithic quantum computer.

```python
display(computer.circuit_diagram())
```

Perform some one-qubit and two-qubit operations (gates) on the qubits in the monolithic quantum computer.

```python
computer.hadamard(0)
computer.controlled_phase(pi / 8, 0, 1)
computer.controlled_phase(pi / 4, 0, 2)
computer.swap(0, 3)
```

Display the circuit with the added gates. Here we can see that the operations on the logical qubits are simply mapped one-to-one on the underlying physical qubits. For a clustered quantum computer, this mapping is more complicated.

```python
display(computer.circuit_diagram())
```

Run the circuit, using number 5 (binary 0101) as the input value:

```python
computer.run(input_number=5)
```

Display the circuit again, showing the initialization:

```python
display(computer.circuit_diagram(with_input=True))
```

```python
computer.statevector_latex()
```

```python
computer.bloch_multivector()
```
