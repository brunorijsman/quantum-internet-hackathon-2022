```python
from numpy import pi
from qiskit import QuantumCircuit, Aer
from qiskit.quantum_info import DensityMatrix
from qiskit.visualization import plot_bloch_multivector, plot_state_city
from qiskit_textbook.tools import array_to_latex
```

```python
def show_state(qc):
    display(qc.draw())
    sim = Aer.get_backend("aer_simulator")
    qc.save_statevector()
    result = statevector = sim.run(qc).result()
    statevector = result.get_statevector()
    display(plot_state_city(statevector))
    density_matrix = DensityMatrix(statevector)
    print("Density matrix = ", density_matrix.data)
    display(plot_bloch_multivector(statevector))
    array_to_latex(statevector, pretext="\\text{Statevector} = ")
    None
```

Density matrix for state $\ket{0}$


```python
qc = QuantumCircuit(1)
show_state(qc)
```

Density matrix for state $\ket{+}$

```python
qc = QuantumCircuit(1)
qc.h(0)
show_state(qc)
```

Density matrix for state $\ket{\Phi^{+}}$

```python
qc = QuantumCircuit(2)
qc.h(0)
qc.cnot(0, 1)
show_state(qc)
```
